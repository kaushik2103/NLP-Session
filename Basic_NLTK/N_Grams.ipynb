```python
# Basic libraries
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
import nltk
```

```python
nltk.download('punkt')
# Tokenization of a paragraph
sentenaces = nltk.sent_tokenize(paragraph)
```

```python
print(sentenaces)
```

```python
type(sentenaces)
```

```python
stemmer = PorterStemmer()
stemmer.stem("running")
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
lemmatizer = WordNetLemmatizer()
lemmatizer.lemmatize("goes")
```

```python
# Want to clean this entire text
import re
corpus = [] # after cleaning
for i in range(len(sentenaces)):
    review = re.sub('[^a-zA-Z]', ' ', sentenaces[i])
    review = review.lower()
    corpus.append(review)

corpus
```

```python
nltk.download('stopwords')
```

```python
# Stemming
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(stemmer.stem(word))
```

```python
# Lammatization
for i in corpus:
    words = nltk.word_tokenize(i)
    for word in words:
        if word not in set(stopwords.words('english')):
            print(lemmatizer.lemmatize(word))
```

```python
# Applying stopwords. Lammatization gives better results
import re
corpus = [] # after cleaning
for i in range(len(sentenaces)):
    review = re.sub('[^a-zA-Z]', ' ', sentenaces[i])
    review = review.lower()
    review = review.split()
    review = [lemmatizer.lemmatize(word) for word in review if not word in set(stopwords.words('english'))]
    review = ' '.join(review)
    corpus.append(review)

corpus
```

```python
# Bag of words
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(binary=False, ngram_range=(2,3))
```

```python
X = cv.fit_transform(corpus)
X
```

```python
cv.vocabulary_
```

```python
X.toarray()
```

```python
## TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=100)
X = tfidf.fit_transform(corpus)
X
```
